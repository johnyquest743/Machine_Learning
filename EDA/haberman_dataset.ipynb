```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

iris=pd.read_csv('./data/iris.csv')
hbds=pd.read_csv('./data/haberman.csv')
```

```python
hbds.plot(kind='scatter', x='age', y='nodes')
plt.show()
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(hbds, hue='status' , size=7)\
    .map(plt.scatter, x='nodes', y='age').add_legend()
plt.show()
```

```python
plt.close()
sns.set_style('whitegrid')
sns.pairplot(hbds, hue='status', height=3)
plt.show()
```
